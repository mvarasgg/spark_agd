```python
# 2. Cree dos tablas SparkSQL y almacene el csv en su correspondiente tabla.
# Observación: tenga especial cuidado con los encabezados de los archivos CSV.
# Usted puede considerar cambiar los encabezados de los CSV originales
# Tu código a continuación...
context_sql.sql('DROP TABLE IF EXISTS asistencia')
context_sql.sql('DROP TABLE IF EXISTS evaluacion')
context_sql.sql("""
CREATE TABLE
asistencia (
    marca STRING,
    nombre STRING,
    pertenece STRING,
    expone STRING
)
USING CSV OPTIONS ( 
    header='true',
    nullvalue='NA',
    path='asistencia.csv')
"""
);
context_sql.sql("""
CREATE TABLE
evaluacion (
    marca STRING,
    nombre STRING,
    a_evaluar STRING,
    introduccion INT,
    equipo INT,
    problema INT,
    ventajas INT,
    solucion INT,
    producto INT,
    traccion INT,
    mercado INT,
    competencia INT,
    modelo INT,
    inversion INT,
    contacto INT,
    coordinacion INT,
    claridad INT,
    diapositivas INT,
    a_invertir STRING,
    observaciones STRING
)
USING com.databricks.spark.csv 
OPTIONS (
    path 'evaluacion.csv', 
    header 'true'
)
"""
);
```

```python
# 2. Consulte el listado total de estudiantes (correos electrónicos) del 
# curso de Ingeniería de Software, ordenados alfabéticamente
# Tu código a continuación..
df = context_sql.sql("select nombre from asistencia order by nombre")
df.show()
```

```python
# 3. Consulte la cantidad de asistencias registradas por estudiante; además, la fecha y hora de la primera asistencia
# y la fecha y hora de la última asistencia
# Tu código a continuación...
df = context_sql.sql("SELECT nombre, COUNT(marca), MIN(marca) as primera_asistencia, MAX(marca) as ultima_asistencia FROM asistencia GROUP By nombre")
df.show()
```

```python
# 4. Consulte el listado de estudiantes que asistieron a 2 presentaciones o menos (una).
# Tu código a continuación...
df = context_sql.sql("""
    SELECT nombre
    FROM asistencia
    GROUP By nombre
    HAVING COUNT(marca) <= 2
""")
df.show()
```

```python
# 5. Consulte el listado de estudiantes que no asistieron a ninguna presentación.
# Tu código a continuación...
df = context_sql.sql("""
    SELECT nombre
    FROM asistencia
    GROUP By nombre
    HAVING COUNT(marca) = 0
""")
df.show()
```

```python
# 6. Consulte los integrantes por cada equipo al que pertenecen.
# Tu código a continuación...
df = context_sql.sql("""
    SELECT replace(pertenece, 'Pertenezco a: ', ''), collect_list(nombre)
    FROM asistencia
    GROUP By pertenece
""")
df.show()
```

```python
# 7. Consulte la cantidad de asistentes por presentación, sin considerar los asistentes que pertenecen 
# al equipo que realizó la presentación.
# Tu código a continuación...
df = context_sql.sql("""
    SELECT COUNT(nombre), replace(expone, 'Expone: ', '') as expone
    FROM asistencia
    WHERE replace(pertenece, 'Pertenezco a: ', '') != replace(expone, 'Expone: ', '')
    GROUP BY expone
""")
df.show()
```

```python
# 8. Consutar cuáles integrantes evaluaron a su propio equipo. Estas evaluaciones no serán válidas, pues un
# integrante no puede evaluar a su propio equipo.
# Tu código a continuación...
df = context_sql.sql("""
    SELECT nombre
    FROM asistencia
    WHERE replace(pertenece, 'Pertenezco a: ', '') == replace(expone, 'Expone: ', '')
    GROUP BY nombre
""")
df.show()
```

```python
# 9. Consultar la nota promedio por cada ítem (1 al 15), y la nota promedio total del cada equipo. Recuerde que no
# son válidas las evaluaciones realizadas por los miembros del mismo equipo.
# Tu código a continuación...
df = context_sql.sql("""
    SELECT
        replace(expone, 'Expone: ', '') as equipo,
        AVG((
            introduccion +
            equipo +
            problema +
            ventajas +
            solucion +
            producto +
            traccion +
            mercado +
            competencia +
            modelo +
            inversion +
            contacto +
            coordinacion +
            claridad +
            diapositivas
        ) / 15) as media,
        AVG(introduccion) as c1,
        AVG(equipo) as c2,
        AVG(problema) as c3,
        AVG(ventajas) as c4,
        AVG(solucion) as c5,
        AVG(producto) as c6,
        AVG(traccion) as c7,
        AVG(mercado) as c8,
        AVG(competencia) as c9,
        AVG(modelo) as c10,
        AVG(inversion) as c11,
        AVG(contacto) as c12,
        AVG(coordinacion) as c13,
        AVG(claridad) as c14,
        AVG(diapositivas) as c15
    FROM evaluacion JOIN asistencia ON evaluacion.nombre = asistencia.nombre
    WHERE replace(pertenece, 'Pertenezco a: ', '') != replace(expone, 'Expone: ', '')
    GROUP BY expone
""")
df.show()
```
